```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la página
url = 'https://edgar.jrc.ec.europa.eu/report_2020'

# Realizar la solicitud HTTP
response = requests.get(url)

# Parsear el contenido HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Encontrar todas las tablas en la página
tables = soup.find_all('table')

# Lista para almacenar los DataFrames de todas las tablas
all_tables_dataframes = []

# Recorrer todas las tablas encontradas
for table in tables:
    # Extraer los encabezados de la tabla (si los hay)
    headers = [header.text.strip() for header in table.find_all('th')]

    # Extraer las filas de la tabla
    rows = []
    for row in table.find_all('tr')[1:]:  # Ignorar la primera fila de encabezados
        columns = row.find_all('td')
        if columns:
            row_data = [col.text.strip() for col in columns]

            # Verificar si la fila tiene el mismo número de columnas que los encabezados
            if len(row_data) == len(headers):
                rows.append(row_data)
            else:
                # Rellenar con valores vacíos si no coincide el número de columnas
                while len(row_data) < len(headers):
                    row_data.append('')
                rows.append(row_data)

    # Si hay filas extraídas, convertirlas en un DataFrame
    if rows:
        df = pd.DataFrame(rows, columns=headers)
        all_tables_dataframes.append(df)
df
# Mostrar los DataFrames generados (uno por cada tabla)
# for index, df in enumerate(all_tables_dataframes):
#     print(f"Tabla {index + 1}:")
#     print(df)
#     print("\n")
```
