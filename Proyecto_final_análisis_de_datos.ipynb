```python
mediageneracionenergiaEolica = (df[df["tipo"]=="Eólico"])
mediaeolica = mediageneracionenergiaEolica["energia_kwh_d_a"].mean()
mediacostoseolica = mediageneracionenergiaEolica["inversi_n_estimada_cop"].mean()
valoreolica = mediaeolica/mediacostoseolica


mediageneracionenergiasolar	 = (df[df["tipo"]=="Solar"])
mediasolar = mediageneracionenergiasolar	["energia_kwh_d_a"].mean()
mediacostosolar = mediageneracionenergiasolar	["inversi_n_estimada_cop"].mean()
valorsolar = mediasolar/mediacostosolar

resumen = {
    "media eolica": [mediaeolica,  mediacostoseolica, valoreolica],
    "media solar": [mediasolar,  mediacostosolar, valorsolar]
}

df1 = pd.DataFrame(resumen)
print(df1)


```

En la tabla anterior, la fila 1 muestra la media de la cantidad de energía generada por cada tipo (eólica y solar). La fila 2 presenta la media del costo de producción por cada tipo de energía. Finalmente, en la fila 3, se muestra la división entre la media de la cantidad de energía y la media del costo de producción. Este cálculo nos da como resultado el costo promedio de producción de cada tipo de energía.

```python
import pandas as pd
import matplotlib.pyplot as plt
df1.plot(kind='bar', figsize=(10, 6))
plt.title('Comparación entre Energía Eólica y Solar', fontsize=14)
plt.ylabel('Valor', fontsize=12)
plt.xlabel('Tipo de Energía', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

En la gráfica se observa que la producción de energía eólica es significativamente más costosa que la de energía solar. Un kWh de energía eólica tuvo un costo de 2.58 centavos, mientras que un kWh de energía solar costó 1.24 centavos. Esto significa que producir energía eólica es 2.08 veces más costoso que producir energía solar. Cabe destacar que 1 kWh es una unidad de energía equivalente a 1000 vatios-hora.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression

# Cargar el DataFrame de energías renovables
data = "https://www.datos.gov.co/resource/vy9n-w6hc.json?$query=SELECT%20capacidad%2C%20proyecto%2C%20tipo%2C%20departamento%2C%20municipio%2C%20codigo_departamento%2C%20codigo_municipio%2C%20fecha_estimada_fpo%2C%20energia_kwh_d_a%2C%20usuarios%2C%20inversi_n_estimada_cop%2C%20empleos_estimados%2C%20emisiones_co2_ton_a_o"
df = pd.read_json(data)

# Limpiar y preparar el DataFrame de energías renovables
df['fecha_estimada_fpo'] = df['fecha_estimada_fpo'].str.replace('T00:00:00.000', '', regex=False)

# Intentar convertir las fechas, y manejar los errores (fechas fuera de rango se convierten en NaT)
df['fecha_estimada_fpo'] = pd.to_datetime(df['fecha_estimada_fpo'], errors='coerce')

# Eliminar filas donde la fecha es NaT (fuera de rango o inválidas)
df = df.dropna(subset=['fecha_estimada_fpo'])

# Extraer el año de las fechas
df['Año'] = df['fecha_estimada_fpo'].dt.year

# Sumarizar la generación de energía por año
df_energy = df.groupby('Año')['energia_kwh_d_a'].sum().reset_index()

# Cargar el DataFrame de emisiones de CO2
url = 'https://edgar.jrc.ec.europa.eu/booklet/EDGARv5.0_FT2019_fossil_CO2_booklet2020.xls'
df_co2 = pd.read_excel(url, engine='xlrd', sheet_name='fossil_CO2_per_capita_by_countr')

# Filtrar los datos de Colombia y preparar el DataFrame
colombia = df_co2.iloc[40]
colombia_df = colombia.to_frame().reset_index()
colombia_df.columns = ['Año', 'CO2_emisiones']

# Asegurar que los años coinciden en ambos DataFrames (se pueden unir por el año)
df_merged = pd.merge(df_energy, colombia_df, on='Año', how='inner')

# Verificar si hay valores NaN después de la fusión
print(f"Valores NaN después de la fusión: {df_merged.isnull().sum()}")

# Ver las primeras filas de los datos fusionados
print(df_merged.head())

# Matriz de correlación entre las columnas de interés
df_goal_values = df_merged[['energia_kwh_d_a', 'CO2_emisiones']]  # Seleccionar columnas relevantes para la correlación
matriz_correlacion = df_goal_values.corr()  # Calcular la matriz de correlación

# Mostrar la matriz de correlación como un mapa de calor
plt.figure(figsize=(12, 10))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriz de Correlación entre Generación de Energía Renovable y Emisiones de CO2')
plt.show()
```

Conclusiones sobre la correlación entre la generación de energías renovables y las emisiones de CO2 en Colombia:

1. El valor de 0.604547 indica una **correlación positiva moderada** entre la generación de energía renovable (en kWh/día) y las emisiones de CO2 (en toneladas per cápita) para Colombia.

2. Esto sugiere que, a medida que aumenta la generación de energía renovable, las emisiones de CO2 tienden a aumentar también, pero esta relación no es extremadamente fuerte. La correlación positiva no es lo suficientemente alta como para indicar una relación directa perfecta, pero sí sugiere que existe cierta asociación entre ambas variables.

3. **Posibles razones para la correlación moderada:**
*   Energía renovable y emisiones de CO2: Aunque las energías renovables generalmente tienen un impacto menor en las emisiones de CO2 en comparación con las fuentes de energía fósil, la relación observada podría reflejar un aumento en las emisiones debido a la expansión de proyectos de energía renovable.
*   Construcción e infraestructura: La construcción de nuevas plantas de energía renovable, como parques eólicos o solares, puede generar emisiones de CO2 en su fase de construcción. Estos proyectos tienen una huella de carbono asociada durante su desarrollo, lo que podría explicar la correlación positiva observada.
*   Demanda de energía: Un aumento en la generación de energía renovable puede ir acompañado de un crecimiento en la demanda de energía. Este incremento en la demanda podría derivar en una mayor actividad económica, lo que a su vez puede resultar en más emisiones de CO2 provenientes de sectores no renovables.

4. **Mezcla energética:** La mejora en la generación de energía renovable no siempre se traduce de manera inmediata en una reducción proporcional de las emisiones de CO2. Esto depende de la composición de la matriz energética total del país. Si el país sigue dependiendo de fuentes fósiles para complementar la generación de energía renovable, las emisiones de CO2 pueden seguir siendo altas, a pesar de un incremento en la producción de energías renovables.

5. **Relación a largo plazo:** Es importante tener en cuenta que los efectos de la energía renovable sobre las emisiones de CO2 pueden no ser inmediatos. El impacto real de la energía renovable en la reducción de emisiones podría volverse más evidente con el tiempo, a medida que reemplaza progresivamente fuentes fósiles más contaminantes.
